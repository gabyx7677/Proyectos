```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm
from sklearn.cluster import KMeans

"""
Generador de datos sinteticos
"""
mean_1 = [0, 0]  # Centro de los datos
cov_1 = [[50, 0], [0, 20]]  # Covarianza
mean_2 = [50, 50]
cov_2 = [[50, 0], [0, 50]]
x_1, y_1 = np.random.multivariate_normal(mean_1, cov_1, 100).T
x_2, y_2 = np.random.multivariate_normal(mean_2, cov_2, 100).T
plt.plot(x_1, y_1, 'o')
plt.plot(x_2, y_2, 'x')
"""
Creación de modelo
"""
X = []  # Matriz de entrenamiento
Y = []  # Etiquetas
for i in range(100):
    X.append([x_1[i], y_1[i]])
    Y.append(1)
for i in range(100):
    X.append([x_2[i], y_2[i]])
    Y.append(-1)
"""
Clasificacion de los datos
"""
clf = svm.SVC()
clf.fit(X, Y)  # Metodo para encontrar los vectores de soporte
print("1", clf.predict([[10, 10]]))  # Probar clasificador
"""
Agrupación de los datos
"""
kmeans = KMeans(n_clusters=2, random_state=0).fit(X) # Calcular la agrupación de k-means
print("2", kmeans.labels_)
print("3", kmeans.predict([[0, 0], [50, 50]])) # Probar agrupador
plt.show()
```
